```python
from apify_client import ApifyClient
import json
import pandas as pd
import sqlite3
import json
```

```python
def fetch_data():
    client = ApifyClient("apify_api_VfYMRA9EmVQQVcirL4sjgG9s39Y1AD3UVvwZ")
    hashtags = ['gorpcore']
    
    # Prepare the Actor input
    run_input = {
        "directUrls": [f"https://www.instagram.com/explore/tags/{hashtag}/" for hashtag in hashtags],
        "resultsType": "details",
        "resultsLimit": 200,
        "searchType": "hashtag",
        "searchLimit": 1,
        "extendOutputFunction": """async ({ data, item, helpers, page, customData, label }) => {
      return item;
    }""",
        "extendScraperFunction": """async ({ page, request, label, response, helpers, requestQueue, logins, addProfile, addPost, addLocation, addHashtag, doRequest, customData, Apify }) => {
     
    }""",
        "customData": {},
    }
    
    # Run the actor
    run = client.actor("apify/instagram-scraper").call(run_input=run_input)
    items = []
    
    # Fetch actor results from the run's dataset
    for item in client.dataset(run["defaultDatasetId"]).iterate_items():
        # Get top_posts list as it's a nested dict
        top_posts = item.get('topPosts', [])
        for top in top_posts:
            items.append(top)
    
    return items
```

```python
def insert_db(data):
    # Connect to the SQLite database
    conn = sqlite3.connect('insta_hashtag_data.db')
    cur = conn.cursor()

    # Iterate over each item in the data and insert it into the database
    for item in data:
        values = (
            item['id'],
            item['type'],
            item['shortCode'], 
            item['caption'],
            json.dumps(item['hashtags']),  # Updated to match the key name in your data
            json.dumps(item['mentions']),  # Updated to match the key name in your data
            item['url'],
            item['commentsCount'],  # Updated to match the key name in your data
            item['dimensionsHeight'],  # Updated to match the key name in your data
            item['dimensionsWidth'],  # Updated to match the key name in your data
            item['displayUrl'],  # Updated to match the key name in your data
            json.dumps(item['images']),  # Updated to match the key name in your data
            item['alt'],
            item['likesCount'],  # Updated to match the key name in your data
            item['timestamp'],
            item['ownerId'],  # Updated to match the key name in your data
        )
        
        # Define a SQL query to insert the data into the database
        insert_query = """
        INSERT OR IGNORE INTO insta_hashtag_data (
            id, type, shortCode, caption, hashtags, mentions, url, commentsCount, 
            dimensionsHeight, dimensionsWidth, 
            displayUrl, images, alt, likesCount, timestamp, ownerId
        )
        VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?);
        """
        
        # Execute the SQL query to insert the data
        cur.execute(insert_query, values)

    # Commit the changes to save the data to the database
    conn.commit()

    # Close the connection
    conn.close()
```

```python
def main():
    
    new_data = fetch_data()

    insert_db(new_data)

    print("Data pipeline executed successfully.")


if __name__ == "__main__":
    main()
```
